```python
df1=pd.read_csv("laptopPrice.csv")
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(data=df1,x='Number of Ratings',y='Number of Reviews')
plt.title("First Graph")
plt.show()
```

### IQR MIN=Q1-1.5IQR MAX=Q3+1.5IQR

```python
cat = list(df1.columns[df1.dtypes=="object"])
cat
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(data=df1,x="brand",y="Price")
plt.title("Boxplot")
plt.show()
```

```python
for i in cat:
    plt.figure(figsize=(10,5))
    sns.boxplot(data=df1,x=i,y="Price")
    plt.title(f"Boxplot for {i} and Price")
    plt.show()
```

```python
### Show the average price of laptops by brand
```

```python
b=df1.groupby(by='brand').agg({'Price':'mean'}).sort_values(by='Price')
```

```python
b.plot(kind='barh')
```

```python
### Get median price by processor brand and processor name
```

```python
c2=df1.groupby(by=["processor_brand","processor_name"]).agg({'Price':'median'})
```

```python
c2
```

```python
c2.plot(kind='barh')
```
